**Question 2: Write a Python program to reverse a string without using [::-1]**

```python
st = "DataScience"
rev = ""

for i in range(len(st)) :
  rev += st[(len(st)-1)-i]
print(rev)
```

**Question 3: Write a Python program to check if a string is a palindrome don't use [::-1].**

```python
st = "racecar"
rev = ""

for i in range(len(st)) :
  rev += st[(len(st)-1)-i]
if rev == st :
  print("Given String is a Palindrome")
else :
  print("Given String is not a Palindrome")

```

**Question 5: Write a Python program to check if two strings are anagrams.**

```python
var1 = 'listen'
var2 = 'silent'
cou = 0

for i in var2 :
  if i in var1 :
    cou += 1
if cou == len(var2) :
  print("Both are anagrams")
else :
  print("Both are not anagrams")
```

**Question 6: Write a Python program to remove all the vowels from a string.**

```python
st = "datascience"
v_st = "aeiouAEIOU"
res = ""
vowels = ""

for i in st :
  if i in v_st :
    vowels += i
  else :
    res += i
print(res)
print(vowels)
```

**Question 7: Write a Python program to find the longest word in a string.**


```python
st = 'python is Easy Language for DataScience'
sl = st.split(" ")
max_co = len(sl[0])
for i in sl :
  c = len(i)
  if c > max_co :
    max_co = c
for i in sl :
  if len(i) == max_co :
    print(i)
```

**Question 8: Write a Python program to capitalize the first letter of each word in a string [dont use built-in-function called title].**

```python
st = 'python is easy'
sl = st.split(' ')
res_l = []
for i in sl :
  j = list(i)
  j[0] = j[0].upper()
  k = "".join(j)
  res_l.append(k)
res_string = " ".join(res_l)
print(res_string)
```

**Question 9: Write a Python program to find the frequency of each character in a string.**


```python
var = 'sharuk khan'

d = {}
for i in var :
  c = 0
  for j in var :
    if i == j :
      c += 1
  d[i] = c
print(d)
```

**Question 10: write a python programme to find the sum of all the even characters based on Ascii values**

```python
var = 'sharuk khan'
even_sum = 0
for i in var :
  val = ord(i)
  if val % 2 == 0 :
    even_sum += val
print(even_sum)
```
